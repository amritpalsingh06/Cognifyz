## Task: Table Booking and Online Delivery

```python
import pandas as pd
```

```python
data = pd.read_csv("Dataset .csv")
data
```

### The percentage of restaurants that offer table booking and online delivery.

```python
table_booking_percentage = data['Has Table booking'].value_counts(normalize = True)*100
print(f"Percentage of table booking :" , table_booking_percentage)
```

```python
online_delivery_percentage = data['Has Online delivery'].value_counts(normalize = True)*100
print(f"Percentage of online delivery :" , online_delivery_percentage)
```

### The average ratings of restaurants with table booking and those without.

```python
avg_rating_with_table_booking = data[data['Has Table booking'] == "Yes"]['Aggregate rating'].mean()
print(f"Average rating with table booking :" , avg_rating_with_table_booking)
```

```python
avg_rating_without_table_booking = data[data['Has Table booking'] == "No"]['Aggregate rating'].mean()
print(f"Average rating without table booking :" , avg_rating_without_table_booking)
```

### The availability of online delivery among restaurants with different price ranges.

```python
online_delivery_by_price_range = data.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack() * 100
print("Percentage of restaurants offering online delivery by price range :\n" , online_delivery_by_price_range)
```

## Task: Price Range Analysis

### Determining the most common price range among all the restaurants.

```python
most_common_price_range = data['Price range'].mode().iloc[0]
print(f"Most Common Price Range :", most_common_price_range)
```

### Calculating the average rating for each price range.

```python
average_rating_by_price_range = data.groupby('Price range')['Votes'].mean()
print("Average Rating by Price Range :\n" , average_rating_by_price_range)
```

### Identifying the color that represents the highest average rating among different price ranges.

```python
highest_rating_color = average_rating_by_price_range.idxmax()
print(f"Color representing the highest average rating :" , highest_rating_color)
```

## Task: Feature Engineering

### Extracting additional features from the existing columns, such as the length of the restaurant name or address.

```python
data['Restaurant Name Length'] = data['Restaurant Name'].apply(len)
```

```python
data[['Restaurant Name Length']]
```

### Create new features like "Has Table Booking" or "Has Online Delivery" by encoding categorical variables.

```python
data['Has Table booking'] = data['Has Table booking'].map({True: 1, False: 0})
data['Has Online delivery'] = data['Has Online delivery'].map({True: 1, False: 0})
```
